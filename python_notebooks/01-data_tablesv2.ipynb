```python
# Helper functions
def look_up_response_meanings(label):
    print(label)
    if main_lut.loc[label,'type'] in ['c','mc','mco']:
        return lut.loc[label,['name','response']].set_index('name')
    else:
        return lut.loc[main_lut.loc[label,'type'], ['name','response']].set_index('name')

def index_with_meaning(df,label):
    new_indexes = df.index.astype(int).astype(str)
    if main_lut.loc[label,'type'] in ['c','mc','mco']:
        pass
    else:
        response_meanings = lut.loc[main_lut.loc[label,'type'], ['name','response']].set_index('name')
    return response_meanings.loc[new_indexes,'response']
```

```python
df = pd.read_csv('vars_of_interest_v2.csv',index_col=0)
```

```python
df['hh_size'] = main_df.hh_size
```

```python
df.loc[df.hh_size == 0,'hh_size'] = 1
```

```python
df['AVG_INCOME_USD_PERCAP'] = df['AVG_INCOME_USD'] / df['hh_size']
```

```python
df['REMESA_AMOUNT_REGPERMONTH_PERCAP'] = df['REMESA_AMOUNT_REGPERMONTH'] / df['hh_size']
```

```python
df.loc[:,[col for col in df.columns if 'pct_remit_towards' in col ]] = df.loc[:,[col for col in df.columns if 'pct_remit_towards' in col ]].fillna(0)
```
